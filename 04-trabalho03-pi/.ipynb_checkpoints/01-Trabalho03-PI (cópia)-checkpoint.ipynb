### Passo 01: Importar bibliotecas

```python
%matplotlib inline
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

### Passo 02: Ler imagem

```python
image = cv.imread('input_images/baboon.png',0)
```

### Passo 03: Gerar funções auxiliares

```python
#Fills the mask in a circular way with the given factor value
def fillCircMask(mask,factor,radius,center,top,bottom,left,right):
    
    for line in range(top, bottom):
        for col in range(left, right):
            if((center[0] - line)**2 + (center[1] - col)**2) <= radius**2:
                mask[line, col] = factor
```

```python
#Generates Low, High and Band Pass Ideal Kernels
def genCircIdealKernel(image_shape, radius, ftype='LPF', small_radius = 0): #Cite reference
    
    center = (image_shape[0]//2, image_shape[1]//2)
    
    top_limit = center[0] - radius
    bottom_limit = center[0] + radius + 1
    left_limit = center[1] - radius
    right_limit = center[1] + radius + 1
     
    if ftype == 'LPF':
        mask = np.zeros((image_shape[0],image_shape[1],2),np.uint8)
        factor = 255
    elif ftype == 'HPF':
        mask = np.ones((image_shape[0],image_shape[1],2),np.uint8)*255
        factor = 0
    elif ftype == 'BPF':
        mask = np.zeros((image_shape[0],image_shape[1],2),np.uint8)
        small_mask = np.ones((image_shape[0],image_shape[1],2),np.uint8)
        factor = 255
        small_factor = 0
        
        fillCircMask(mask,factor,radius,center,top_limit,bottom_limit,left_limit,right_limit)
        fillCircMask(small_mask, small_factor,small_radius,center,top_limit,bottom_limit,left_limit,right_limit)
        mask = mask * small_mask
        
        return mask
    
    fillCircMask(mask,factor,radius,center,top_limit,bottom_limit,left_limit,right_limit)
    
    return mask
```

### Passo 04: Gerar Kernels

```python
lpf_ideal_kernel = genCircIdealKernel(image.shape,radius=1, ftype='LPF')
```

```python
hpf_ideal_kernel = genCircIdealKernel(image.shape,radius=50, ftype='HPF')
```

```python
bpf_ideal_kernel = genCircIdealKernel(image.shape,radius=50, ftype='BPF', small_radius=20)
```

### Passo 05: Gerar espectros de Fourier

```python
dft_image = cv.dft(np.float32(image),flags = cv.DFT_COMPLEX_OUTPUT) # Generates complex array
dft_image = np.fft.fftshift(dft_image) # Shifts the spectrum to the image center
```

```python
magnitude_image = 20*np.log(cv.magnitude(dft_image[:,:,0],dft_image[:,:,1])) #Generates the magnitude image
```

```python
plt.subplot(121),plt.imshow(image, cmap = 'gray',vmin=0, vmax=255)
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(magnitude_image , cmap = 'gray', vmin=0, vmax=255)
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

### Passo 06: Aplicar filtros

```python
lpf_spectrum = dft_image * (lpf_ideal_kernel/255)
hpf_spectrum = dft_image * (hpf_ideal_kernel/255)
bpf_spectrum = dft_image * (bpf_ideal_kernel/255)
```

### Passo 07: Gerar imagens filtradas

```python
#Shifts back the spectrum to the top left
lpf_image = np.fft.ifftshift(lpf_spectrum)
hpf_image = np.fft.ifftshift(hpf_spectrum)
bpf_image = np.fft.ifftshift(bpf_spectrum)
```

```python
#Applies the inverse transform
lpf_image = cv.idft(lpf_image)
hpf_image = cv.idft(hpf_image)
bpf_image = cv.idft(bpf_image)
```

```python
#Gets the real part of the image
lpf_image = cv.magnitude(lpf_image[:,:,0],lpf_image[:,:,1])
hpf_image = cv.magnitude(hpf_image[:,:,0],hpf_image[:,:,1])
bpf_image = cv.magnitude(bpf_image[:,:,0],bpf_image[:,:,1])
```

```python
#Convert from Float64 to Int8
lpf_image = ((lpf_image/np.max(lpf_image))*255).astype(np.uint8)
hpf_image = ((hpf_image/np.max(hpf_image))*255).astype(np.uint8)
bpf_image = ((bpf_image/np.max(bpf_image))*255).astype(np.uint8)
```

### Passo 08: Mostrar Imagens

```python
plt.figure(figsize=(10,10))
plt.subplot(221),plt.imshow(image, cmap = 'gray',vmin=0, vmax=255)
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(222),plt.imshow(lpf_image, cmap = 'gray')
plt.title('Low Pass Resulting Image'), plt.xticks([]), plt.yticks([])
plt.subplot(223),plt.imshow(hpf_image, cmap = 'gray')
plt.title('High Pass Resulting Image'), plt.xticks([]), plt.yticks([])
plt.subplot(224),plt.imshow(bpf_image, cmap = 'gray')
plt.title('Band Pass Resulting Image'), plt.xticks([]), plt.yticks([])
plt.show()
```

### Passo 09: Compressão das imagens

```python
import math
```

```python
dfft_copy = dft_image.copy()
```

```python
def getIndexesMag(mag, limit):
    indexes = []
    for line in range(mag.shape[0]):
        for col in range(mag.shape[1]):
            if mag[line,col] < limit:
                indexes.append((line, col))
    return indexes
```

```python
for line in range(dfft_copy.shape[0]):
    for col in range(dfft_copy.shape[1]):
        if math.sqrt(dfft_copy[line,col,0]**2 + dfft_copy[line,col,1]**2) < 200:
            dfft_copy[line,col] = 0
```
