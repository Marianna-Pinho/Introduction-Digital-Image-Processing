### Passo 01: Importar bibliotecas

```python
%matplotlib inline
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
```

### Passo 02: Criar funções auxiliares

```python
def plotImage(image):
    plt.figure(figsize=(5, 5))
    plt.imshow(image, cmap='gray', vmin=0, vmax=255)
    plt.show()
```

```python
def plotAllImages(images):
    columns = 5 
    rows = 2
    i = 1
    fig=plt.figure(figsize=(50, 50))

    for key in images.keys():
        fig.add_subplot(rows, columns, i)
        plt.xlabel(str(key))
        plt.imshow(images[key], cmap = 'gray', vmin=0, vmax=255)
        i = i+1

    plt.subplots_adjust(bottom=0.40, top=0.75)
    plt.show()
```

### Passo 03: Criar Filtros

#### Filtro h1: Filtro Laplaciano com centro positivo (passa-alta) 5x5 

```python
filter_h1 = np.array([[0,0,-1,0,0],[0,-1,-2,-1,0],[-1,-2,16,-2,-1],[0,-1,-2,-1,0],[0,0,-1,0,0]])
filter_h1
```

#### Filtro h2: Filtro Gaussiano

```python
filter_h2 = np.array([[1,4,6,4,1],[4,16,24,16,4],[6,24,36,24,6],[4,16,24,16,4],[1,4,6,4,1]])
h2_nf = 1/256

print(h2_nf)
print(filter_h2)
```

#### Filtro h3: Operador de Sobel usado para filtro de gradiente (gy)

```python
filter_h3 = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
filter_h3
```

#### Filtro h4: Operador de Sobel usado para filtro de gradiente (gx)

```python
filter_h4 = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
filter_h4
```

#### Filtro h5: filtro laplaciano com centro positivo (passa alta) 3x3

```python
filter_h5 = np.array([[-1,-1,-1],[-1,8,-1],[-1,-1,-1]])
filter_h5
```

#### Filtro h6: filtro da média (passa baixa)

```python
filter_h6 = np.array([[1,1,1],[1,1,1],[1,1,1]])
h6_nf = 1/9

print(h6_nf)
print(filter_h6)
```

#### Filtro h7

#### Combinação dos filtros h3 e h4

```python
filtered_images['image_h3_h4_2'] = np.sqrt(np.power(filtered_images['image_h3']/255, 2) + np.power(filtered_images['image_h4']/255,2))
filtered_images['image_h3_h4_2'] = (filtered_images['image_h3_h4_2']/np.max(filtered_images['image_h3_h4_2']))*255
```

**2 - Tentar fazer funcionar o último item sem normalizar a imagem de entrada. O problema, provavelmente, está no tipo de dados.** <br>

A imagem de entrada precisa ser normalizada pelos mesmos motivos que normalizamos a imagem de entrada no trabalho 1, quando realizamos a operação de exponenciação quadrada: ao elevarmos os valores originais, eles se tornam grandes, não cabendo no tipo de dados adotado. 

Outra solução que poderia ser adotada é a conversão do tipo de dados da imagem de entrada, de uint8 para float64. Assim, os valores não ultrapassariam o limite máximo permitido pelo tipo de dados e os resultados seriam corretos. Entretanto, entre expandir o tipo de dados e normalizar a imagem de entrada, a segunda opção parece mais atrativa. Principalmente ao pensarmos em consumo de memória.
