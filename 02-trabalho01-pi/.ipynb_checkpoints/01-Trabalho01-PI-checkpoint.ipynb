## Importar Bibliotecas

```python
%matplotlib inline
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
from skimage.exposure import rescale_intensity
```

## Questão 01 - Resolução de Imagens

### Passo 01: Definir funções auxiliares

```python
''' Using max(1,image.shape//factor) we can keep decreasing the largest dimension, when using rectangular images,
once the smallest will be rounded to 1.'''

def changeResolution(image, factor):
    new_image = cv.resize(image,(max(1,image.shape[1]//factor), max(1,image.shape[0]//factor)), interpolation = cv.INTER_AREA)
    return cv.resize(new_image, (image.shape[1],image.shape[0]),interpolation = cv.INTER_AREA)
```

### Passo 02: Ler imagem

```python
image = cv.imread('input_images/baboon.png',0)
```

### Passo 03: Alterar resolução das imagens

```python
line_shape = image.shape[0]
col_shape = image.shape[1]
```

```python
rescaled_images = {}
max_size = max(line_shape, col_shape)
size_range = [1 * 2 ** (n) for n in range(1, int(np.log2(max_size)) + 1)]

for k in size_range:
    rescaled_images[max_size//k] = changeResolution(image,k)
```

### Passo 04: Apresentar as imagens

```python
columns = 5 
rows = 3
i = 1
fig=plt.figure(figsize=(30, 30))

for key in rescaled_images.keys():
    fig.add_subplot(rows, columns, i)
    plt.xlabel('Resolution: ' + str(key) + ' x ' + str(key) + ' pixels')
    plt.imshow(rescaled_images[key], cmap = 'gray', vmin=0, vmax=255)
    i = i+1
    
plt.subplots_adjust(bottom=0.40, top=0.75)
plt.show()
```

### Passo 05: Salvar as imagens

```python
for key in rescaled_images.keys():
    cv.imwrite('output_images/q1_normal_baboon_'+str(key)+'.png', rescaled_images[key])
```

## Questão 02 - Quantização de Imagens

### Passo 01: Definir funções auxiliares

```python
def quantizeImage(image, level):
    factor = 256//level
    new_image = (image//factor)*factor
    return rescale_intensity(new_image)
```

### Passo 02: Ler imagem

```python
image = cv.imread('input_images/baboon.png',0)
```

### Passo 03: Quantizar imagem

```python
quantized_images = {}
max_level = 256
level_range = [1 * 2 ** (n) for n in range(1, int(np.log2(max_level)) + 1)]
```

```python
for level in level_range:
    quantized_images[level] = quantizeImage(image,level)
```

### Passo 04: Mostrar imagens

```python
columns = 4 
rows = 2
i = 1
fig=plt.figure(figsize=(30, 30))

for key in quantized_images.keys():
    fig.add_subplot(rows, columns, i)
    plt.xlabel('Niveis de cinza: '+str(key))
    plt.imshow(quantized_images[key], cmap = 'gray', vmin=0, vmax=255)
    i = i+1
    
plt.subplots_adjust(bottom=0.40, top=0.75)
plt.show()
```

### Passo 05: Salvar as imagens

```python
for key in quantized_images.keys():
    cv.imwrite('output_images/q2_level_' + str(key) + '.png', quantized_images[key])
```

## Questão 03 - Escala de Cinza

### Passo 01: Definir funções pedidas

```python
def transLogarithm(image):
    max_image = np.max(abs(image)) #in case of negative numbers
    if(max_image == 0): #to prevent division by zero
        alpha = 0
    else:
        alpha = 255/np.log(1 + max_image)
    return alpha * np.log(image + 1)
```

```python
def transExponential(image):
    new_image = image/np.max(image) #normalize image
    max_intensity = np.max(new_image)
    
    if(max_intensity > 0): #to prevent division by zero
        alpha = 255/(np.exp(max_intensity) - 1)
    else:
        alpha = 255/np.exp(max_intensity)
        
    return alpha * (np.exp(new_image) - 1)
```

```python
def transPower2(image):
    new_image = image/np.max(image)
    alpha = 255/np.power(np.max(new_image),2)
    return alpha*np.power(new_image,2)
```

```python
def transSquareRoot(image):
    alpha = 255/np.sqrt(np.max(image))
    return alpha*np.sqrt(image)
```

```python
def transEnlargeContrast(image, alfa, beta, gama, a, b, L):
    new_image = np.where(image >= 0, np.where(image <= a, image*alfa,
                              np.where(image <= b, beta*(image - a) + alfa*a,
                                       np.where(image <= L, gama*(image - b) + beta*(b-a) + alfa*a, 
                                                image))),image)
    return (new_image/np.max(new_image))*255
```

### Passo 02: Ler imagem

```python
image = cv.imread('input_images/butterfly.png',0)
```

### Passo 03: Aplicar transformações

```python
log_image = transLogarithm(image.copy()).astype(np.uint8)
```

```python
exp_image = transExponential(image.copy()).astype(np.uint8)
```

```python
power2_image = transPower2(image.copy()).astype(np.uint8)
```

```python
root_image = transSquareRoot(image.copy()).astype(np.uint8)
```

```python
enlarge_contrast_image = transEnlargeContrast(image.copy(),0.3,1,2,45,190,255).astype(np.uint8)
```

### Passo 04: Apresentar imagens

```python
transformed_images = {}
transformed_images['original'] = image
transformed_images['log_image'] = log_image
transformed_images['exp_image'] = exp_image
transformed_images['power2_image'] = power2_image
transformed_images['root_image'] = root_image
transformed_images['enlarge_contrast_image'] = enlarge_contrast_image
```

```python
columns = 3 
rows = 2
i = 1
fig=plt.figure(figsize=(30, 30))

for key in transformed_images.keys():
    fig.add_subplot(rows, columns, i)
    plt.xlabel('Transformação: '+str(key))
    plt.imshow(transformed_images[key], cmap = 'gray', vmin=0, vmax=255)
    i = i+1
    
plt.subplots_adjust(bottom=0.40, top=0.75)
plt.show()
```

### Passo 05: Salvar imagens

```python
for key in transformed_images.keys():
    if 'original' == str(key):
        continue
    cv.imwrite('output_images/q3_'+str(key)+'.png', transformed_images[key])
```
