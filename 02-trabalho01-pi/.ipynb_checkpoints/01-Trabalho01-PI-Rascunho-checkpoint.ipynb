## Questão 01 - Resolução de Imagens

```python
image = cv.imread('input_images/baboon.png',0)
```

```python
#Colocando max(1, new_shape) garante que não tentaremos fazer o resize da imagem para uma dimensão 0, uma vez que
#ela precisa ser, pelo menos, 1. Também, garantimos que poderemos continuar a diminuir a dimensão maior, até que ela
#também chegue a 1. Isso, talvez, cubra os casos em que as dimensões são diferentes (retângulos).

def resize_image(image, factor):
    return cv.resize(image,(max(1,image.shape[0]//factor), max(1,image.shape[1]//factor)), interpolation = cv.INTER_AREA)
```

```python
line_shape = image.shape[0]
col_shape = image.shape[1]
```

```python
k = 1
rescaled_images = {}
size = max(line_shape, col_shape)

while k <= size:
    rescaled_images[(size//k)] = resize_image(image, k)
    print('factor:{}'.format(k))
    k = k*2
```

```python
columns = 4 
rows = 3
i = 1
fig=plt.figure(figsize=(30, 30))

for key in rescaled_images.keys():
    fig.add_subplot(rows, columns, i)
    plt.xlabel('Resolução '+str(int(key))+' x ' + str(int(key)) + ' pixels')
    plt.imshow(rescaled_images[key], cmap = 'gray', vmin=0, vmax=255)
    i = i+1
    
plt.subplots_adjust(bottom=0.40, top=0.75)
plt.show()
```

```python
def rebin( a, newshape ):
        '''Rebin an array to a new shape.
        '''
        assert len(a.shape) == len(newshape)

        slices = [ slice(0,old, float(old)/new) for old,new in zip(a.shape,newshape) ]
        coordinates = np.mgrid[slices]
        indices = coordinates.astype('i')   #choose the biggest smaller integer index
        return a[tuple(indices)]
```

## Questão 02 - Quantização de Imagens

```python
image = cv.imread('input_images/baboon.png',0)
```

```python
def quantizeImage(image, levels):
    factor = 256//levels
    new_image = (image//factor)*factor
    return skimage.exposure.rescale_intensity(new_image) #puts the image in the 0 ... 255 brightness scale
```

```python
newi = quantizeImage(image, 4)
newi
```

### Logaritmo

```python
def transformLog(image):
    alpha = 255/np.log(1 + np.max(image))
    return alpha * np.log(image + 1)
```

```python
log_image = transformLog(image).astype(np.uint8)
log_image
```

```python
cv.imshow('log',log_image)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Exponencial

```python
image = cv.imread('input_images/butterfly.png',0)
```

```python
def transformExp(image):  #alpha = 255/(np.exp(np.max(image)) - 1)
    new_image = image/255
    alpha = 255/np.exp(np.max(new_image))
    return alpha * (np.exp(new_image) - 1)
```

```python
def transformExp2(image):
    #Calculates e^image
    down_scale_image = image/255
    #print("Down scale image\n", down_scale_image)
    
    down_exp_image = np.exp(down_scale_image)
    #print("Down exp image\n", down_exp_image)
    
    exp_image = np.power(down_exp_image, 255) - 1
    #print("Exp image",exp_image)
    
    #Calculates the maximum value of exp_image
    exp_max_intensity = np.max(exp_image)
    #print(exp_max_intensity)
    
    #Puts image intensities in the [0,255 interval]
    normalized_image = (exp_image - np.min(exp_image))/(np.max(exp_image) - np.min(exp_image))
    #print("Normalized image", normalized_image)
    
    up_scaled_image = normalized_image*255
    return up_scaled_image
```

```python
exp_image = transformExp(image).astype(np.uint8)
```

```python
cv.imshow('exp',exp_image)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Quadrado 

```python
def transformSquare(image):
    new_image = image/255
    alpha = 255/np.power(np.max(new_image),2)
    return alpha*np.power(new_image,2)
```

```python
square_image = transformSquare(image).astype(np.uint8)
```

```python
cv.imshow('square',square_image)
cv.waitKey(0)
cv.destroyAllWindows()
```

### Raiz quadrada

```python
def transformRoot(image):
    alpha = 255/np.sqrt(np.max(image))
    return alpha*np.sqrt(image)
```

```python
root_image = transformRoot(image).astype(np.uint8)
```

```python
cv.imshow('root',root_image)
cv.waitKey(0)
cv.destroyAllWindows()
```

## Alargamento de contraste

```python
image = cv.imread('input_images/butterfly.png',0)
```

```python
def transformEnlarge(image, alfa, beta, gama, a, b, L):
    lines = image.shape[0]
    cols = image.shape[1]
    
    for line in range(0,lines):
        for col in range(1,cols):
            val = image[line, col]
            if((0 <= val) and (val <= a)):
                image[line,col] = alfa*val
            elif((a < val) and (val <= b)):
                image[line,col] = beta*(val - a) + alfa*a
            elif((b < val) and (val <= L)):
                image[line,col] = gama*(val - b) + beta*(b-a) + alfa*a
    return image
```

```python
enlarge_image = transformEnlarge(image.copy(),1,3,0.5,60,180,255)
enlarge_image
```

```python
cv.imshow('enlarge',enlarge_image)
cv.waitKey(0)
cv.destroyAllWindows()
```
